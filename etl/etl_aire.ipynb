También, la columna "time_period" la voy a dividir en otras dos columnas, por un lado, una va a contener el año, y otra la información adicional de la fecha

```python
# Función para extraer el año
def extract_year(text):
    import re
    match = re.search(r'\b\d{4}\b', text)
    return match.group(0) if match else None

# Función para extraer la descripción
def extract_description(text):
    import re
    match = re.search(r'\b\d{4}\b', text)
    if match:
        return text.replace(match.group(0), '').strip()
    else:
        return text

# Aplicar las funciones al DataFrame
aire['año'] = aire['time_period'].apply(extract_year)
aire['descripcion_temporal'] = aire['time_period'].apply(extract_description)

# Llenar valores nulos en la columna 'description' con una cadena vacía
aire['descripcion_temporal'] = aire['descripcion_temporal'].fillna('')
```
