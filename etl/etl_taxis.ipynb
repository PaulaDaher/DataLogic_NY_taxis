Ahora lo que voy a hacer es primero acomodar las columnas de manera que nos queden las columnas más importantes primero y las secundarias a lo último. Según nuestro análisis:

- Importantes:
    - tpep_pickup_datetime

```python
# Ordenando columnas...
taxis_enero_2024 = taxis_enero_2024[['tpep_pickup_datetime', 'tpep_dropoff_datetime', 
       'trip_distance', 'PULocationID', 'DOLocationID', 'total_amount', 'passenger_count', 'payment_type']]
# Muestro en pantalla las primeras 3 filas 
taxis_enero_2024.head(3)
```

Ahora voy a filtrar por aquellos PULocationID y DOLocationID, que únicamente hayan sido en Manhattan. Para ello primero importamos otro dataset...

```python
zona_taxis = pd.read_csv("../dataset/taxi+_zone_lookup.csv")
```

```python
zona_taxis
```

```python
# Filtramos únicamente dónde "Borough" = Manhattan
zona_taxis[zona_taxis["Borough"] == "Manhattan"]
```

```python
Manhattan = zona_taxis[zona_taxis["Borough"] == "Manhattan"] # Hago un dataframe que contenga únicamente los datos de Manhattan
```

De este último Dataframe, se podría eliminar la columna "Borough" porque se vuelve redundante

```python
Manhattan = Manhattan.drop("Borough",axis=1)
```

Ahora, voy a filtrar el dataframe de taxis por aquellas filas dónde: "PULocationID" sea igual a "LocationID"

```python
# Lista que va a contener las ID correspondientes a las zonas de Manhattan
lista_de_ID_zonas_Manhattan = []
# Bucle que itera cada ID de la columna de Manhattan
for id in Manhattan["LocationID"]:
    # En cada iteración, se agrega la id a la lista
    lista_de_ID_zonas_Manhattan.append(id)
```

Aplicamos con una máscara un filtro al Dataframe, y lo mostramos en pantalla

```python
taxis_enero_2024[(taxis_enero_2024["PULocationID"].isin(lista_de_ID_zonas_Manhattan)) & (taxis_enero_2024["DOLocationID"].isin(lista_de_ID_zonas_Manhattan))]
```

Guardamos el dataframe filtrado en una variable local

```python
registros_taxis_enero_2024_en_manhattan = taxis_enero_2024[(taxis_enero_2024["PULocationID"].isin(lista_de_ID_zonas_Manhattan)) & (taxis_enero_2024["DOLocationID"].isin(lista_de_ID_zonas_Manhattan))]
```

Ahora, agregamos la columna "zona_subida" y "zona_bajada" al dataframe principal, basandonos en la columna "zone" del dataframe Manhattan uniendo ambas tablas.

```python
# Realizamos el primer merge
registros_ampliado_por_zona_subida = pd.merge(registros_taxis_enero_2024_en_manhattan, Manhattan, left_on='PULocationID', right_on='LocationID', how='left')

# Hacemos el segundo merge
registros_taxis_enero_2024_en_manhattan = pd.merge(registros_ampliado_por_zona_subida, Manhattan, left_on='DOLocationID', right_on='LocationID', how='left')

# Mostramos en pantalla
registros_taxis_enero_2024_en_manhattan
```

Ahora, notemos que "Zone_x" corresponde con nuestra zona de subida, ya que coinciden las columnas "PULocationID" con "LocationID_x", y que "Zone_y" corresponde con nuestra zona de bajada. Para hacer esto más claro, vamos a eliminar aquellas columnas que no nos son útiles, para posteriormente renombrar las columnas de las zonas, de una forma más clara

```python
registros_taxis_enero_2024_en_manhattan.drop(columns=["LocationID_x","service_zone_x","LocationID_y","service_zone_y"])
```

```python
registros_taxis_enero_2024_en_manhattan = registros_taxis_enero_2024_en_manhattan.drop(columns=["LocationID_x","service_zone_x","LocationID_y","service_zone_y"])
```

Con estos cambios, vamos a reestructurar nuevamente el orden de las columnas para finalmente renombrarlas y terminar con nuestro trabajo

```python
registros_taxis_enero_2024_en_manhattan.columns
```

```python
registros_taxis_enero_2024_en_manhattan = registros_taxis_enero_2024_en_manhattan[['tpep_pickup_datetime', 'Zone_x', 'tpep_dropoff_datetime', 'Zone_y', 'trip_distance', 'total_amount',
       'PULocationID', 'DOLocationID', 'passenger_count',
       'payment_type']]
```

```python
registros_taxis_enero_2024_en_manhattan
```

Por último, cambiamos los nombres de las columnas por otros un tanto más descriptivos y en español

```python
diccionario_de_renombre = {'tpep_pickup_datetime':"fecha_hora_subida", 'Zone_x':"zona_subida", 'tpep_dropoff_datetime':"fecha_hora_bajada", 'Zone_y':"zona_bajada", 'trip_distance':"distancia_recorrida", 'total_amount':"monto_total",
       'PULocationID':"id_zona_subida", 'DOLocationID':"id_zona_bajada", 'passenger_count':"cantidad_pasajeros",
       'payment_type':"tipo_pago"}
```

```python
registros_taxis_enero_2024_en_manhattan.rename(columns=diccionario_de_renombre)
```

Guardamos el dataframe final :)

```python
registros_taxis_enero_2024_en_manhattan = registros_taxis_enero_2024_en_manhattan.rename(columns=diccionario_de_renombre)
```
