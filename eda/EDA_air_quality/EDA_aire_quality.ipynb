También, la columna "Time Period" la voy a dividir en otras dos columnas, por un lado, una va a contener el año, y otra la información adicional de la fecha

```python
# Función para extraer el año
def extract_year(text):
    import re
    match = re.search(r'\b\d{4}\b', text)
    return match.group(0) if match else None

# Función para extraer la descripción
def extract_description(text):
    import re
    match = re.search(r'\b\d{4}\b', text)
    if match:
        return text.replace(match.group(0), '').strip()
    else:
        return text

# Aplicar las funciones al DataFrame
quality['año'] = quality['Time Period'].apply(extract_year)
quality['Descripcion Temporal'] = quality['Time Period'].apply(extract_description)

# Llenar valores nulos en la columna 'description' con una cadena vacía
quality['Descripcion Temporal'] = quality['Descripcion Temporal'].fillna('')
```

```python
#creacion de lista con los barrios de manhattan
barrios_manhattan = [
    "Washington Heights", "Inwood", "Marble Hill", "Central Harlem", "East Harlem",
    "Upper East Side", "Yorkville", "Lenox Hill", "Roosevelt Island", "Upper West Side",
    "Lincoln Square", "Clinton", "Midtown", "Murray Hill", "Turtle Bay", "Gramercy",
    "Stuyvesant Town", "Chelsea", "Greenwich Village", "SoHo", "Lower East Side",
    "Tribeca", "Little Italy", "Chinatown", "Battery Park City", "Financial District"
]
```

```python
# Filtrar el dataframe para incluir solo los barrios de Manhattan
df_manhattan = year_quality[year_quality['nombre_del_barrio'].isin(barrios_manhattan)]
```

## CREACION DE GRAFICOS

### Histograma de Niveles de Contaminación

```python
#Histograma de 'valor_del_dato'
plt.figure(figsize=(9, 6))
sns.histplot(df_manhattan['valor_del_dato'], kde=True)
plt.title('Distribución de Los niveles de Contaminacion en Manhattan')
plt.xlabel('Niveles de Contaminación')
plt.ylabel('Frecuencia')
plt.show()
```

### Serie temporal

```python
# Serie temporal
df_manhattan['fecha_inicial'] = pd.to_datetime(df_manhattan['fecha_inicial'])

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_manhattan, x='fecha_inicial', y='valor_del_dato', hue='nombre_del_barrio')
plt.title('Tendencia Temporal de Niveles de Contaminación en Manhattan por Barrio')
plt.xlabel('Periodo de los años')
plt.ylabel('Niveles de Contaminación')
plt.legend(title='Barrio', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

### Comparación Geográfica

```python
#Comparación Geográfica
plt.figure(figsize=(14, 7))
sns.boxplot(data=df_manhattan, x='nombre_del_barrio', y='valor_del_dato')
plt.title('Comparación de Niveles de Contaminación en Manhattan por Barrio')
plt.xlabel('Barrio')
plt.ylabel('Niveles de Contaminación')
plt.xticks(rotation=90)
plt.show()
```

### Promedio de Niveles de Contaminación por Año y Barrio

```python
#Promedio de Niveles de Contaminación por Año y Barrio
df_grouped_manhattan = df_manhattan.groupby(['año', 'nombre_del_barrio'])['valor_del_dato'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_grouped_manhattan, x='año', y='valor_del_dato', hue='nombre_del_barrio')
plt.title('Promedio de Niveles de Contaminación en Manhattan por Año y Barrio')
plt.xlabel('Año')
plt.ylabel('Niveles de Contaminación')
plt.legend(title='Barrio', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
