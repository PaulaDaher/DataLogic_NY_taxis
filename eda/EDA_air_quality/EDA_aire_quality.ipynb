TambiÃ©n, la columna "Time Period" la voy a dividir en otras dos columnas, por un lado, una va a contener el aÃ±o, y otra la informaciÃ³n adicional de la fecha

```python
# FunciÃ³n para extraer el aÃ±o
def extract_year(text):
    import re
    match = re.search(r'\b\d{4}\b', text)
    return match.group(0) if match else None

# FunciÃ³n para extraer la descripciÃ³n
def extract_description(text):
    import re
    match = re.search(r'\b\d{4}\b', text)
    if match:
        return text.replace(match.group(0), '').strip()
    else:
        return text

# Aplicar las funciones al DataFrame
quality['aÃ±o'] = quality['Time Period'].apply(extract_year)
quality['Descripcion Temporal'] = quality['Time Period'].apply(extract_description)

# Llenar valores nulos en la columna 'description' con una cadena vacÃ­a
quality['Descripcion Temporal'] = quality['Descripcion Temporal'].fillna('')
```

### Disminucion del tamaÃ±o del dataset, filtrando la columna 'aÃ±o'

```python
#creacion de lista con los barrios de manhattan
barrios_manhattan = [
    "Washington Heights", "Inwood", "Marble Hill", "Central Harlem", "East Harlem",
    "Upper East Side", "Yorkville", "Lenox Hill", "Roosevelt Island", "Upper West Side",
    "Lincoln Square", "Clinton", "Midtown", "Murray Hill", "Turtle Bay", "Gramercy",
    "Stuyvesant Town", "Chelsea", "Greenwich Village", "SoHo", "Lower East Side",
    "Tribeca", "Little Italy", "Chinatown", "Battery Park City", "Financial District"
]
```

```python
# Filtrar el dataframe para incluir solo los barrios de Manhattan
df_manhattan = year_quality[year_quality['nombre_del_barrio'].isin(barrios_manhattan)]
```

## CREACION DE GRAFICOS

### Histograma de Niveles de ContaminaciÃ³n

```python
#Histograma de 'valor_del_dato'
plt.figure(figsize=(9, 6))
sns.histplot(df_manhattan['valor_del_dato'], kde=True)
plt.title('DistribuciÃ³n de Los niveles de Contaminacion en Manhattan')
plt.xlabel('Niveles de ContaminaciÃ³n')
plt.ylabel('Frecuencia')
plt.show()
```

### Serie temporal

```python
# Serie temporal
df_manhattan['fecha_inicial'] = pd.to_datetime(df_manhattan['fecha_inicial'])

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_manhattan, x='fecha_inicial', y='valor_del_dato', hue='nombre_del_barrio')
plt.title('Tendencia Temporal de Niveles de ContaminaciÃ³n en Manhattan por Barrio')
plt.xlabel('Periodo de los aÃ±os')
plt.ylabel('Niveles de ContaminaciÃ³n')
plt.legend(title='Barrio', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

### ComparaciÃ³n GeogrÃ¡fica

```python
#ComparaciÃ³n GeogrÃ¡fica
plt.figure(figsize=(14, 7))
sns.boxplot(data=df_manhattan, x='nombre_del_barrio', y='valor_del_dato')
plt.title('ComparaciÃ³n de Niveles de ContaminaciÃ³n en Manhattan por Barrio')
plt.xlabel('Barrio')
plt.ylabel('Niveles de ContaminaciÃ³n')
plt.xticks(rotation=90)
plt.show()
```

### Promedio de Niveles de ContaminaciÃ³n por AÃ±o y Barrio

```python
#Promedio de Niveles de ContaminaciÃ³n por AÃ±o y Barrio
df_grouped_manhattan = df_manhattan.groupby(['aÃ±o', 'nombre_del_barrio'])['valor_del_dato'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_grouped_manhattan, x='aÃ±o', y='valor_del_dato', hue='nombre_del_barrio')
plt.title('Promedio de Niveles de ContaminaciÃ³n en Manhattan por AÃ±o y Barrio')
plt.xlabel('AÃ±o')
plt.ylabel('Niveles de ContaminaciÃ³n')
plt.legend(title='Barrio', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
