```python
#Trabajaremos solo con las marcas que nos interesan para el analisis
# Lista de las marcas que queremos filtrar
marcas_filtradas = ['Chrysler','Ford','Mercedes-Benz','Nissan','Toyota','Hyundai','Kia','Tesla','Toyota','Honda']
# Filtrar el DataFrame
df_filtrado_marcas = df_electric[df_electric['Make'].isin(marcas_filtradas)]
```

```python
df_filtrado_marcas
```

```python
#Vemos los modelos de autos que fabrican las marcas que nos interesan
df_filtrado_marcas['Model'].unique()
```

Modelos aceptados para ser utilizados como taxis > Pacifica  |  Voyager | Transit | Metris | NV200 | Sienna | Mustang | Ioniq | Niro | Model 3 | Model Y | bZ4X | Escape | Accord | Niro | Avalon | Camry | Highlander | Prius | RAV4 | Sienna | Venza | Escape | Altima | Camry | Crown

```python
# Lista de palabras clave
palabras_clave = [
    'Pacifica', 'Voyager', 'Transit', 'Metris', 'NV200', 'Sienna', 'Mustang',
    'Ioniq', 'Niro', 'Model 3', 'Model Y', 'bZ4X', 'Escape', 'Accord', 'Avalon',
    'Camry', 'Highlander', 'Prius', 'RAV4', 'Venza', 'Altima', 'Crown'
]

# Crear una expresión regular que coincida con cualquiera de las palabras clave
expresion_regular = '|'.join(palabras_clave)

# Filtrar el DataFrame
df_filtrado_modelos = df_filtrado_marcas[df_filtrado_marcas['Model'].str.contains(expresion_regular, case=False, na=False)]
```

```python
df_filtrado_modelos
```

```python
#Analizamos el dataset resultante 
df_filtrado_modelos.info()
```

```python
df_filtrado_modelos.shape
```

```python
#eliminamos columnas que son irrelevantes para el analisis 
df_electric2 = df_filtrado_modelos.drop(columns=['CO2 rating', 'Smog rating'], axis=1)
df_electric2.head()
```

```python
#Renombramos las columnas 
df_electric2.rename(columns={'Model year': 'model_year', 'Make': 'Brand', 'Vehicle class': 'vehicle_class', 'Engine size (L)': 'engine_size_L', 'Cylinders':'engine_cylinder', 'Transmission':'transmission_type', 'Fuel type':'vehicle_type'}, inplace=True)
```

```python
df_electric2.head(2)
```

```python
# Reordenamos las columnas para que coincidan con los otros datasets 
df_electric3 = df_electric2.reindex(columns=['Brand', 'Model', 'model_year','vehicle_class','vehicle_type', 'engine_size_L','engine_cylinder','transmission_type','City (L/100 km)','Highway (L/100 km)','Combined (L/100 km)','Combined (mpg)','CO2 emissions (g/km)'])
```

```python
df_electric3.head()
```

```python
df_electric3['Model'].unique()
```

```python
df_electric3['vehicle_class'].unique()
```

```python
#Remplazamos los valores de la columna vehicle class para unificar los valores 
df_electric3['vehicle_class'] = df_electric3['vehicle_class'].replace({'Sport utility vehicle: Small': 'SUV', 'Sport utility vehicle: Standard': 'SUV'})
```

```python

df_electric3['vehicle_type'] = df_electric3['vehicle_type'].replace({'X': 'Electric'})
```

```python
#Los modelos hybridos los identificamos con la palabra Hybrid
df_electric3.loc[df_electric3['Model'].str.contains('Hybrid', case=False, na=False), 'vehicle_type'] = 'Hybrid'
```

```python
# eliminamos columnas irrelevantes para nuestro analisis
df_electric4 = df_electric3.drop(columns=['City (L/100 km)', 'Highway (L/100 km)','Combined (L/100 km)'], axis=1)
df_electric4.head()
```
