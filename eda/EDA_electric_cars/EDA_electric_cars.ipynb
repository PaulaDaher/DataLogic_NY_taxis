```python
### Importamos las librer√≠as necesarias

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
### Leemos los datasets

df_electric = pd.read_csv('df_electric_cars.csv')
df_electric_hybrid = pd.read_csv('df_electric_hybrid_cars.csv')
```

### Comenzamos a analizar el comportamiento de los datos 

```python
df = df_electric
```

Analisis de precios 

```python
plt.figure(figsize=(8, 3))
sns.barplot(x='Model', y='price_dollars', data=df, palette='bright', hue='Brand')
plt.xticks(rotation=45)
plt.title('Precio por Modelo')
plt.xlabel('Modelo')
plt.ylabel('Precio en d√≥lares')
plt.show()
```

An√°lisis de Autonomm√≠a

```python
# Gr√°fico de dispersi√≥n de autonom√≠a (range_km) versus precio

plt.figure(figsize=(10, 6))
sns.scatterplot(x='range_km', y='price_dollars', hue='Brand', data=df, palette='bright', s=200)
plt.title('Autonom√≠a versus Precio')
plt.xlabel('Autonom√≠a (km)')
plt.ylabel('Precio en d√≥lares')
plt.show()
```

An√°lisis de Consumo de Wh por km recorrido

```python
# 4. Gr√°fico de l√≠neas del consumo combinado por modelo
plt.figure(figsize=(10, 6))
sns.lineplot(x='Model', y='Combined (wh/km)', data=df, marker='o')
plt.xticks(rotation=45)
plt.title('Consumo Combinado por Modelo')
plt.xlabel('Modelo')
plt.ylabel('Consumo Combinado (wh/km)')
plt.show()
```

```python
# Ordenar los datos seg√∫n la autonom√≠a de mayor a menor
df_sorted = df.sort_values(by='range_km', ascending=False)

# Crear el gr√°fico de barras
plt.figure(figsize=(12, 3))
sns.barplot(x='range_km', y='Model', data=df_sorted, palette='viridis', hue='Brand')
plt.title('Modelos ordenados por autonom√≠a (de mayor a menor)')
plt.xlabel('Autonom√≠a (km)')
plt.ylabel('Modelo')
plt.show()
```

An√°lisis de tiempo de carga 

```python
# Ordenar los datos seg√∫n la autonom√≠a de mayor a menor
df_sorted = df.sort_values(by='charge_time_minutes', ascending=False)

# Crear el gr√°fico de barras
plt.figure(figsize=(12, 3))
sns.barplot(x='charge_time_minutes', y='Model', data=df_sorted, palette='viridis', hue='Brand')
plt.title('Modelos ordenados por carga m√≠nima necesaria')
plt.xlabel('Tiempo de carga m√≠nima')
plt.ylabel('Modelo')
plt.show()
```
