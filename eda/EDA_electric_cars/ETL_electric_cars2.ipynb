```python
# Lista de las marcas que queremos filtrar
marcas_filtradas = ['Chrysler ','Ford ','Mercedes-Benz ','Nissan ','Toyota ','Hyundai ','Kia ','Tesla ','Toyota ','Honda ']
# Filtrar el DataFrame
df_filtrado_marcas = df_electric[df_electric['Brand'].isin(marcas_filtradas)]
```

```python
df_filtrado_marcas
```

```python
#Filtramos por modelos 

df_filtrado_marcas['Model'].unique()
```

Modelos que se aceptan para taxis > Pacifica  |  Voyager | Transit | Metris | NV200 | Sienna | Mustang | Ioniq | Niro | Model 3 | Model Y | bZ4X | Escape | Accord | Niro | Avalon | Camry | Highlander | Prius | RAV4 | Sienna | Venza | Escape | Altima | Camry | Crown

```python
#filtramos el dataset por modelos de autos 
# Lista de palabras clave
palabras_clave = [
    'Pacifica', 'Voyager', 'Transit', 'Metris', 'NV200', 'Sienna', 'Mustang',
    'Ioniq', 'Niro', 'Model 3', 'Model Y', 'bZ4X', 'Escape', 'Accord', 'Avalon',
    'Camry', 'Highlander', 'Prius', 'RAV4', 'Venza', 'Altima', 'Crown'
]

# Crear una expresiÃ³n regular que coincida con cualquiera de las palabras clave
expresion_regular = '|'.join(palabras_clave)

# Filtrar el DataFrame
df_filtrado_modelos = df_filtrado_marcas[df_filtrado_marcas['Model'].str.contains(expresion_regular, case=False, na=False)]
```

```python
df_filtrado_modelos
```

```python
#Eliminamos columnas que no necesitamos 
df_electric2 = df_filtrado_modelos.drop(columns=['AccelSec', 'TopSpeed_KmH', 'Efficiency_WhKm', 'FastCharge_KmH','RapidCharge', 'PowerTrain','Segment','Seats'], axis=1)
```

```python
df_electric2
```

```python
df_electric2['model_year'] = 2024
```

```python
df_electric2.info()
```

```python
#Renombramos las columnas 
df_electric2.rename(columns={'Range_Km': 'range_km', 'PlugType': 'plug_type', 'BodyStyle': 'vehicle_class', 'PriceEuro': 'price_euro'}, inplace=True)
df_electric2
```

```python
#agregamos columanas para poder mergear con otros datasets
df_electric2['vehicle_type'] = 'Electric'
```

```python
#Reordenamos columnas 
df_electric3 = df_electric2.reindex(columns=['Brand', 'Model', 'model_year','vehicle_class','vehicle_type','range_km', 'plug_type', 'price_euro'])
```
