En esta parte del proyecto uniremos los 3 datasets que contienen registros de autos el'ectricos previamente tratados para poder crear un solo dataset de anélisis de modelos de autos electricos e hybridos y sus características técnicas.
El dataset resultante será un dataset que nos servirá para la alimentación del modelo de machine learninf de recomendación de modelos de autos 

```python
import pandas as pd
```

```python
#Leemos los csv que luego vamos a unir

df_electric_1 = pd.read_csv('df_electric_1.csv')
df_electric_2 = pd.read_csv('df_electric_2.csv')
df_electric_3 = pd.read_csv('df_electric_3.csv')
```

```python
#concatenamos los tres datasets
df_electric= pd.concat([df_electric_1, df_electric_2, df_electric_3], ignore_index=True, sort=False)
```
