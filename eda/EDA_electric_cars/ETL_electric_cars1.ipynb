```python
#Vemos todos los fabricantes de autos que contiene el CSV
df_electric['Manufacturer'].unique()
```

Marcas que fabrican autos aptos para taxis> Chrysler | Ford | Mercedes-Benz | Nissan | Toyota | Hyundai | Kia | Tesla | Toyota | Honda

```python
# Lista de las marcas que queremos filtrar
marcas_filtradas = ['Chrysler','Ford','Mercedes-Benz','Nissan','Toyota','Hyundai','Kia','Tesla','Toyota','Honda']
# Filtrar el DataFrame con las marcas de la lista
df_filtrado_marcas = df_electric[df_electric['Manufacturer'].isin(marcas_filtradas)]
```

```python
df_filtrado_marcas
```

```python
df_filtrado_marcas['Model'].unique()
```

MODELOS ACEPTADOS PARA SER TAXI> Pacifica  |  Voyager | Transit | Metris | NV200 | Sienna | Mustang | Ioniq | Niro | Model 3 | Model Y | bZ4X | Escape | Accord | Niro | Avalon | Camry | Highlander | Prius | RAV4 | Sienna | Venza | Escape | Altima | Camry | Crown

```python
#Filtramos por modelos aceptados

# Lista de palabras clave
palabras_clave = [
    'Pacifica', 'Voyager', 'Transit Connect', 'Metris', 'NV200', 'Sienna', 'Mustang Mach-E',
    'Ioniq 5', 'Niro', 'Model 3 RWD', 'Model Long Range', 'Model Y Long Range', 'bZ4X', 'Escape', 'Accord', 'Avalon',
    'Camry', 'Highlander', 'Prius', 'RAV4', 'Venza', 'Altima', 'Crown'
]

# Crear una expresión regular que coincida con cualquiera de las palabras clave
expresion_regular = '|'.join(palabras_clave)

# Filtrar el DataFrame
df_filtrado_modelos = df_filtrado_marcas[df_filtrado_marcas['Model'].str.contains(expresion_regular, case=False, na=False)]
```

```python
df_filtrado_modelos
```

```python
#Vemos que el dataset no solo tiene autos electricos e hybridos sino que tambien tiene autos de gasolina

df_filtrado_modelos['Fuel'].unique()
```

Siguiendo el proyecto legislativo de Nueva York quien aspira a que todos los taxis en Nueva York para el 2030 sean eléctricos , eliminaremos de la lista los autos de combustión interna. 

```python
fuels_to_remove = ['Ethanol (E85)', 'Propane - Bi-fuel', 'CNG - Compressed Natural Gas', 'CNG - Bi-fuel', 'Propane']
df_filtrado_fuel = df_filtrado_modelos[~df_filtrado_modelos['Fuel'].isin(fuels_to_remove)]
```

```python
df_filtrado_fuel
```

```python
#Vemos la información del nuevo dataset
df_filtrado_fuel.info()
```

```python

df_electric2 = df_filtrado_fuel.drop(columns=['PHEV Total Range', 'Alternative Fuel Economy City', 'Alternative Fuel Economy Highway', 'Alternative Fuel Economy Combined','Transmission Make', 'Number of Passengers','Heavy-Duty Power System','Notes', 'Drivetrain'], axis=1)
```

```python
#Eliminamos las columnas que no necesitamos 

df_electric3 = df_electric2.drop(columns=['Conventional Fuel Economy City','Conventional Fuel Economy Highway', 'Engine Type'], axis=1)
df_electric3.head()
```

```python
#Renombramos las columnas para que tengan un nombre mas entendible
df_electric3.rename(columns={'Model Year': 'model_year', 'Manufacturer': 'Brand', 'Category': 'vehicle_class', 'Fuel':'vehicle_type', 'All-Electric Range':'range_km', 'Conventional Fuel Economy Combined':'Combined (mpg)','Engine Size': 'engine_size_L', 'Engine Cylinder Count':'engine_cylinder', 'Transmission Type':'transmission_type', 'Fuel type':'vehicle_type'}, inplace=True)
```
