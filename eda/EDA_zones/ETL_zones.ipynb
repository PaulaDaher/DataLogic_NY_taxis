```python
#Importamos librerias necesarias

import pandas as pd
```

```python
# Leemos el dataset

df_zones = pd.read_csv(r'Datasets\taxi_zone_lookup.csv')
df_zones
```

```python
#Filtramos el DF para obtener solo las zonas de Manhattan
df_manhattan = df_zones[df_zones['Borough'] == 'Manhattan']
df_manhattan
```

```python
df_zonas_taxis = df_manhattan.drop('Borough', axis=1, inplace=True)
```

```python
#Exportamos el DF a un archivo CSV
df_zonas_taxis.to_csv('zonas_taxis_en_Manhattan.csv', index = False)
```
