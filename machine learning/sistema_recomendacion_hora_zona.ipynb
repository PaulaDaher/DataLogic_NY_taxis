```python
movimientos_taxis_Manhattan["hora_viaje"] = movimientos_taxis_Manhattan["tpep_pickup_datetime"].dt.hour
movimientos_taxis_Manhattan
```

```python
conteo_hora_zona = movimientos_taxis_Manhattan.groupby(['hora_viaje', 'PULocationID']).size().reset_index(name='conteo')
```

```python
conteo_hora_zona
```

```python
zonas
```

```python
hola = pd.merge(conteo_hora_zona, zonas, left_on="PULocationID", right_on="LocationID", how="left")
```

```python
dataset_con_hora_conteo_zona = hola.drop(columns=["PULocationID","Borough","service_zone"])
```

```python
def top_zonas_por_horario(horario, top_n=5):
    # Filtrar por horario
    if horario >= 0 and horario <= 23:
        df_horario = dataset_con_hora_conteo_zona[dataset_con_hora_conteo_zona['hora_viaje'] == horario]
        # Ordenar por conteo en orden descendente y obtener las top_n zonas
        top_zonas = df_horario.sort_values(by='conteo', ascending=False).head(top_n)
        return top_zonas[["Zone","conteo"]]
    else:
        return f"{horario} no es un hora adecuada. Ingrese otra por favor"
```
